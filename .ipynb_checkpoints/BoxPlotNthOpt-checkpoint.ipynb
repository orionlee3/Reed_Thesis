```python
import numpy as np
import numpy.linalg as npla
import scipy as sp
import joblib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
qr_data_list = np.array([joblib.load("Timing_Data/QR_every_%s.pkl"%(i)) for i in range(1,21)])
lu_data_list = np.array([joblib.load("Timing_Data/LU_every_%s.pkl"%(i)) for i in range(1,21)])
```

```python
df_QR = pd.DataFrame(columns =['%s'%(i) for i in range(1,21)])
df_LU = pd.DataFrame(columns =['%s'%(i) for i in range(1,21)])
```

```python
#df_QR['1'] = joblib.load("Timing_Data/QR.pkl")
#df_LU['1'] = joblib.load("Timing_Data/LU.pkl")

for i,n in enumerate(qr_data_list):
    df_QR['%s'%(i+1)] = n
for i,n in enumerate(lu_data_list):
    df_LU['%s'%(i+1)] = n
```

```python
sns.set(rc={"figure.dpi":300, 'savefig.dpi':600})
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
plot = sns.catplot(data = df_QR,estimator=np.mean, ci = 'sd', capsize = 0.2,kind = 'bar')
plt.xlabel("N Value")
plt.ylabel("Time (s)")
plot.set(ylim = (4,5))
#plt.savefig('BenchmarkingN.pdf')
```

```python
sns.set(rc={"figure.dpi":300, 'savefig.dpi':600})
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
plot = sns.catplot(data = df_LU,estimator=np.mean, ci = 'sd', capsize = 0.2,kind = 'bar')
plt.xlabel("N Value")
plot.set(ylim = (4,5))
plt.ylabel("Time (s)")
```

```python
timing_df = pd.DataFrame(columns = ["LU Average", "LU Std", "QR Average", 'QR Std'])
```

```python
timing_df['LU Average'] = [df_LU['%s'%(i)].mean() for i in range(1,21)]
timing_df['LU Std'] = [df_LU['%s'%(i)].std() for i in range(1,21)]
timing_df['QR Average'] = [df_QR['%s'%(i)].mean() for i in range(1,21)]
timing_df['QR Std'] = [df_QR['%s'%(i)].std() for i in range(1,21)]
```
