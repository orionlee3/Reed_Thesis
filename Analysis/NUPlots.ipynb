```python
import numpy as np
from numpy.linalg import multi_dot
from numpy.linalg import qr
import numpy.linalg as npla
import scipy as sp
from scipy.linalg import block_diag, logm, eigvals
import pandas as pd
import joblib
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.style.use('tableau-colorblind10')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Computer Modern"
})
import matplotlib.colors as mcolors
color = ['b','r','g','purple','orange','y']
```

```python
data_sets = [np.load('Results/Run%s/final_array.npy'%(i)) for i in range(1,6)]
strip_width = [int(len(i[0])/2) for i in data_sets]


# amount of angles = len(data)
# strip width = 1/2 * len(data[0]) 
# strip width = len(data[0]) for CC paper

strip_length = int(1e5)
theta_num = 20
theta_vals = np.linspace(np.pi/15,np.pi/2.4,theta_num)

min_val_full = [[j[i,int((strip_width[n]/ 2) -1)] for i in range(theta_num)] for n,j in enumerate(data_sets)]
min_val = [i[9:-1] for i in min_val_full]
```

```python
def fit(x,a,nu,b):
    return a*(x**nu)+b

fits = [curve_fit(fit,[i for i in range(10)],j) for j in min_val]
params = [fits[i][0] for i in range(len(fits))]
covar = [fits[i][1] for i in range(len(fits))]

len(data_sets[1])
```

```python
plt.figure(dpi = 250)

probabilities = [0.5,1,0.8,0.6,0.75]

for i in range(len(probabilities)):
    plt.scatter([i for i in range(10)],min_val[i], s = 1.4, c = color[i], label = r'Probability $ = %s$'%(probabilities[i]))
    plt.plot([i for i in range(10)],fit([i for i in range(10)],*params[i]), 
            lw = 0.3, c = color[i], label = r'$\nu = %s$'%(round(params[i][1],2)))

plt.legend()
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\gamma_{M/2}$')
plt.show()
```

```python
def xi(l,gamma):
    return ((1/(2*l))*gamma)**(-1)
```

```python

plt.figure(dpi = 250)
size = 1.4

for i in range(len(probabilities)):
    plt.scatter([np.arctanh(np.sin(k)) for k in theta_vals],[xi(strip_length,j) for j in min_val_full[i]], 
            s = size, label = r'Probability $ = %s$'%(probabilities[i]), c = color[i], alpha = 0.7)


plt.legend()
plt.xlabel(r'$\theta_{CCM}$')
plt.ylabel(r'$\Lambda_M(\theta)$')
```
